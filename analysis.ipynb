```python
df['average_score'] = (df['math score'] +df['reading score'] +df['writing score']) / 3
df
```

```python
overall_avg = df['average_score'].mean()
overall_avg
```

```python
max_score = df[['math score', 'reading score', 'writing score']].max()
max_score
```

```python
gender_avg = df.groupby('gender')['average_score'].mean()
gender_avg
```

```python
lunch_avg = df.groupby('lunch')['average_score'].mean()
lunch_avg
```

```python
prep_avg = df.groupby('test preparation course')['average_score'].mean()
prep_avg
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
plt.figure()
gender_avg.plot(kind='bar')
plt.title("Average Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.show()
```

```python
plt.figure()
plt.hist(df['math score'], bins=20)
plt.title("Math Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.barplot(data = df, x='gender',y='math score')
plt.show()

```

```python
sns.displot(data = df, x = 'reading score',bins = 5)
plt.xlabel('Reading Scores')
plt.ylabel('Count')
plt.title('Distribution of Reading scores')
plt.show()
```

```python
sns.displot(data = df, x = 'writing score',bins = 5)
plt.xlabel('Writing score')
plt.ylabel('Count')
plt.title('Distribution of Writing scores')
plt.show()
```

```python
avg_scores = df[['math score','reading score','writing score']].mean()
avg_scores
```

```python
plt.pie(avg_scores.values,labels=avg_scores.index,autopct='%.1f%%')

plt.title("Average Score Distribution")
plt.legend(loc='upper left')
plt.show()
```
